### 3. Using the graph of those variables to find answer to research questions

#### 3.A Plotting the graph

##### 3.A.1 Grpah for Energy vs Valence

```python
ax2 = sns.relplot(
    x = 'Energy', y = "Valence", data = df, col = 'Key', kind = "scatter"
)

ax2.map(plt.axhline, y = 0.5, ls = '--', c = 'red')
ax2.map(plt.axvline, x = 0.5, ls = '--', c = 'red')
ax2.set(xlabel = 'Energy', ylabel = 'Valence')
```

##### 3.A.2 Graph for Tempo vs Valence

```python
ax3 = sns.relplot(
    x = 'Tempo', y = "Valence", col = 'Key', data = df, kind = "scatter"
)
ax3.map(plt.axhline, y = 0.5, ls = '--', c = 'red')
ax3.map(plt.axvline, x = 125, ls = '--', c = 'red')
ax3.set(xlabel = 'Tempo', ylabel = 'Valence')
```
